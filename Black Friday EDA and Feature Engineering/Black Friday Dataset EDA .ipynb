```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline

import warnings
warnings.filterwarnings("ignore")
```

# Problem Statement

A retail company “ABC Private Limited” wants to understand the customer purchase behaviour (specifically, purchase amount) against various products of different categories. They have shared purchase summary of various customers for selected high volume products from last month. The data set also contains customer demographics (age, gender, marital status, city_type, stay_in_current_city), product details (product_id and product category) and Total purchase_amount from last month.

Now, they want to build a model to predict the purchase amount of customer against various products which will help them to create personalized offer for customers against different products.

```python
# importing the dataset

df_train=pd.read_csv('Black Friday Dataset/Black Friday_train.csv')
df_train.head()
```

```python
# import the test data
df_test=pd.read_csv('Black Friday Dataset\Black Friday_test.csv')
df_test.head()
```

```python
# Concatinating both train and test data

df=pd.concat([df_train,df_test])
df.head()
```

```python
df.info()
```

```python
df.describe()
```

```python
# There is no use of User_ID column, drop the User_ID columns 
df.drop(['User_ID'],axis=1,inplace=True) 
```

```python
df.head()
```

```python
# Handling the categorical feature Gender into numeric feature using map method, map method will convert Gender feature into 0,1 numeric feature

df['Gender']=df['Gender'].map({'F':0,'M':1})
df.head()
```

```python
# Handle categorical feature Age
df['Age'].unique()
```

```python
# Handling Age feature using Target guiding (Target ordinal encoding)

df['Age']=df['Age'].map({'0-17':1,'18-25':2,'26-35':3,'36-45':4,'46-50':5,'51-55':6,'55+':7})
df.head()
```

```python
# fixing categorical City_category
df_City=pd.get_dummies(df['City_Category'],drop_first=True,dtype=int)
df_City
```

```python
df_City.head()
```

```python
df=pd.concat([df,df_City],axis=1)
df.head()
```

```python
# drop the City_Category feature
df.drop('City_Category',axis=1,inplace=True)
```

```python
df.head()
```
