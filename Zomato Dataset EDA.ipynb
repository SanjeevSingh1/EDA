```python
# What are the top 10 Cuisines
df_cuisines = df_final.groupby(['Cuisines','Aggregate rating'])['Votes'].sum().sort_values(ascending=False).reset_index()
df_cuisines
```

```python
df_cuisines[:10]
```

```python
plt.figure(figsize=(10,10))
sns.barplot(x='Votes',y='Cuisines',data=df_cuisines[:10])
plt.suptitle('Top 10 Cuisines with respect to Votes', fontsize=15, fontweight='bold', alpha=0.8, y=1)
sns.set_context('talk')
plt.ylabel('Votes')
plt.xlabel('Cuisines')
plt.xticks(rotation=90)
plt.show()
```

# Obervations

1. Italian, American, Pizza Cuisines are most popular and most number of votes and ratings
2. American, Burger, Cafe are second most popular cuisines
3. Continental, American, Asian, North Indian are third most popular cuisines

```python
df_final.columns
```

```python
cuisine_avg_cost = df_final.groupby(['Cuisines','Currency'])['Average Cost for two'].size().sort_values(ascending=False).reset_index()
cuisine_avg_cost
```

```python
cuisine_avg_cost[:10]
```

```python
#Relationships between Cuisines and Avgerage Cost for two

plt.figure(figsize=(15,10))
sns.set_context('talk')
ax=sns.barplot(x='Average Cost for two',y='Cuisines',data=cuisine_avg_cost[:10],palette='hls')
ax.set_title('Relationships between Cuisines and Avgerage Cost for two',size = 20)
ax.set_xlabel('Cost in Rupees')
ax.set_ylabel('')
plt.xticks(rotation=90)
plt.show()
```

# Observation

1. North Indian Cuisines have Average cost for two is 936 Rupees.
2. North Indian, Chinese cuisines have Average cost for two is 511 Rupees.
3. Fast Food and Chinese cuisines have Average cost for two are 348 and 340 Rupees respectively.
4. North Indian, Mughlai cuisines have Average cost for two are 334.
5. Street Food have Average cost for two is 149.
6. North Indian Cuisine have highest average cost for two person and Street Food have lowest average cost for two person.
