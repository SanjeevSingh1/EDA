## Bivariate Analysis:

# Which category is having largest number of installation?

```python
df_cat_installs=df.groupby(['Category'])['Installs'].sum().sort_values(ascending=False).reset_index()
```

```python
df1=df_cat_installs[:10]
```

```python
plt.figure(figsize=(15,10))
sns.set_context('talk')
ax=sns.barplot(x='Installs',y='Category',data=df1,palette='hls')
ax.set_title('Most Popular Categories in PlayStore',size = 20)
ax.set_xlabel('No. of Installations in Billions')
ax.set_ylabel('')
plt.show()
```

Here we can say that 'Game' is the most popular category in PlayStore
